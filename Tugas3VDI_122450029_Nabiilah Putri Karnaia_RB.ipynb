**1. Tujuan Visualisasi**

- Membantu pemangku kepentingan dalam memahami dinamika pasar kerja di Indonesia.

**2. Pengguna**

Pengguna dari visualisasi ini mencakup:
- Pembuat Kebijakan: Mereka yang bertanggung jawab untuk merumuskan kebijakan ketenagakerjaan dan upah. Data ini dapat membantu mereka dalam pengambilan keputusan terkait peraturan upah minimum.
- Peneliti dan Akademisi: Mereka yang melakukan penelitian tentang hubungan antara upah, produktivitas, dan kondisi ekonomi. Data ini bisa menjadi bagian penting dari analisis mereka.
- Perusahaan dan Pengusaha: Pihak yang ingin memahami kondisi pasar tenaga kerja untuk menentukan kebijakan upah dan strategi perekrutan.
- Masyarakat Umum: Individu yang ingin mengetahui tren upah di daerah mereka dan di seluruh Indonesia.

**3. Fungsi Visualisasi**

Fungsi dari visualisasi ini mencakup:
- Edukasi: Memberikan wawasan tentang pergerakan upah pekerja dari tahun ke tahun, memungkinkan pengguna untuk memahami kondisi ekonomi dan daya beli masyarakat.
- Analisis Tren: Membantu dalam menganalisis tren upah di berbagai provinsi, serta perbandingan antar daerah, sehingga dapat diidentifikasi daerah yang mengalami peningkatan atau penurunan upah.
- Dasar untuk Diskusi: Mendorong diskusi di antara pemangku kepentingan, termasuk pembuat kebijakan dan pemimpin industri, tentang perlunya perubahan dalam kebijakan upah dan ketenagakerjaan.

**4. Tone (Nada)**

Nada dari visualisasi ini adalah:
- Informatif dan Formal: Nada yang digunakan harus bersifat formal dan berorientasi pada data, mengingat audiens yang beragam, termasuk pembuat kebijakan dan peneliti.
- Analitis dan Kritis: Memberikan ruang untuk analisis yang mendalam, dengan penekanan pada fakta-fakta yang dapat memengaruhi pengambilan keputusan dan kebijakan upah.

```python
# Data statistik deskriptif
data = {
    'tahun': ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022'],
    'min': [8525, 9940, 10569, 11359, 11828, 12707, 12992, 11734],
    '25%': [11233.5, 12978, 13090, 13235.5, 13820.5, 15894.5, 16085, 15509],
    '50%': [12660, 14068, 14822, 15275, 15849, 18136, 18271, 17006],
    '75%': [14194.5, 16334, 16670.5, 16898.5, 17757.5, 22212.5, 21085.5, 18849.5],
    'max': [19290, 23444, 25657, 25987, 25236, 32138, 30662, 32685]
}

# Ubah dictionary menjadi dataframe
df_stats = pd.DataFrame(data)

# Set ukuran plot
plt.figure(figsize=(10, 6))

# Loop melalui setiap tahun dan membuat boxplot
# Gunakan min, 25%, 50%, 75%, dan max untuk membuat boxplot per tahun
box_data = [df_stats[['min', '25%', '50%', '75%', 'max']].iloc[i].values for i in range(len(df_stats))]

# Membuat boxplot berdasarkan data deskriptif
plt.boxplot(box_data, labels=df_stats['tahun'], showmeans=True)

# Menambahkan judul dan label
plt.title('Distribusi Upah 2015-2022', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Upah (IDR)', fontsize=12)

# Menampilkan plot
plt.grid(True)
plt.show()
```

```python
# Set ukuran plot
plt.figure(figsize=(10, 8))

# Membuat plot bar upah tahun 2022 per provinsi
sns.barplot(x='upah.2022', y='provinsi', data=df, hue='provinsi', palette='viridis', legend=False)

# Menambahkan judul dan label
plt.title('Upah Minimum Tahun 2022 di Setiap Provinsi', fontsize=16)
plt.xlabel('Upah 2022 (IDR)', fontsize=12)
plt.ylabel('Provinsi', fontsize=12)

# Menampilkan plot
plt.show()
```

```python
# Data statistik deskriptif dalam bentuk dictionary
data = {
    'tahun': ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022'],
    'mean': [12928.71, 15085.51, 15669.91, 15893.20, 16629.54, 19485.14, 19384.86, 17891.20],
    'min': [8525, 9940, 10569, 11359, 11828, 12707, 12992, 11734],
    '25%': [11233.5, 12978, 13090, 13235.5, 13820.5, 15894.5, 16085, 15509],
    '50%': [12660, 14068, 14822, 15275, 15849, 18136, 18271, 17006],
    '75%': [14194.5, 16334, 16670.5, 16898.5, 17757.5, 22212.5, 21085.5, 18849.5],
    'max': [19290, 23444, 25657, 25987, 25236, 32138, 30662, 32685]
}

# Konversi ke DataFrame
df_stats = pd.DataFrame(data)

# Set ukuran plot
plt.figure(figsize=(10, 6))

# Plot garis untuk mean, min, dan max
plt.plot(df_stats['tahun'], df_stats['mean'], marker='o', label='Mean')
plt.plot(df_stats['tahun'], df_stats['min'], marker='o', label='Min')
plt.plot(df_stats['tahun'], df_stats['max'], marker='o', label='Max')

# Menambahkan judul dan label
plt.title('Tren Statistik Upah 2015-2022', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Upah (IDR)', fontsize=12)

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
